```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn import linear_model
```

```python
data = pd.read_csv('winequality-red.csv')
```

```python
data.head()
```

```python
len(data)
```

```python
x1 = data[["fixed acidity"]]
y = data[["quality"]]
linear1 = linear_model.LinearRegression()
linear1.fit(x1,y)
y1pred = linear1.predict(x1)
plt.plot(x1,y1pred,color="red")
plt.scatter(x1,y)
plt.xlabel("fixed acidity")
plt.ylabel("quality")
print(linear1.score(x1,y))
```
